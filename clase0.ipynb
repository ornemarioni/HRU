```python
import pyccl as ccl

h1 = 0.674

#cosmologia de Planck II (arXiv:1807.06209)
cosmo1 = ccl.Cosmology(Omega_c=0.264, Omega_b=0.0493, h=0.674, sigma8=0.811, n_s=0.965)

#E-dS
cosmo2 = ccl.Cosmology(Omega_c=1, Omega_b=0, h=0.674, sigma8=0.811, n_s=0.965)

#Low density
cosmo3 = ccl.Cosmology(Omega_c=0.05, Omega_b=0, Omega_k=0.95 ,h=0.674, sigma8=0.811, n_s=0.965)

#High Lambda
cosmo4 = ccl.Cosmology(Omega_c=0.2, Omega_b=0, Omega_k=0 ,h=0.674, sigma8=0.811, n_s=0.965)



c = 300000
H0 = 100.*h1
DH = c/H0
z = np.linspace(0., 5., 200)
a = 1./(z +1.)
```

### Angular diameter distance

```python
DA1 = ccl.comoving_angular_distance(cosmo1,a)
DA2 = ccl.comoving_angular_distance(cosmo2,a)
DA3 = ccl.comoving_angular_distance(cosmo3,a)
DA4 = ccl.comoving_angular_distance(cosmo4,a)
```

```python
# from __future__ import unicode_literals

# matplotlib.rcParams['text.usetex'] = True
# matplotlib.rcParams['text.latex.unicode'] = True

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)


ax.plot(z ,DA2*a/DH,'-', lw=1, color='k', label='E-dS')
ax.plot(z ,DA3*a/DH,':', lw=1, color='k', label='low density')
ax.plot(z ,DA4*a/DH,'--', lw=1, color='k', label='high lambda')
ax.plot(z ,DA1*a/DH,'-.', lw=1, color='k', label='PlanckII')

ax.set_ylabel(r'angular diameter distance $\mathrm{D_A/D_H}$', fontsize=24)
ax.set_xlabel(r'redshift z',fontsize=24)
ax.set_xticks([0,1,2,3,4,5])
# ax.yaxis.set_major_locator(MultipleLocator(0.1))
# ax.yaxis.set_major_formatter(NullFormatter())
# ax.set_yticklabels(['','','$0.1$','','$0.3$','','$0.5$','','$0.7$'])
# ax.axes.ticklabel_format(style='sci')
ax.set_ylim(0,0.5)
ax.set_xlim(0,5)
ax.minorticks_on()
ax.tick_params( labelsize=24)
ax.tick_params('both', length=5, width=1.8,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=8, width=1.8,which='major', direction='in', right='on',top='on')

ax.legend(fontsize=21,frameon=False, loc=4)

plt.show()
```

### Proper motion distance

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)


ax.plot(z ,DA2/DH,'-', lw=1, color='k', label='E-dS')
ax.plot(z ,DA3/DH,':', lw=1, color='k', label='low density')
ax.plot(z ,DA4/DH,'--', lw=1, color='k', label='high lambda')
ax.plot(z ,DA1/DH,'-.', lw=1, color='k', label='PlanckII')

ax.set_ylabel(r'proper motion distance $\mathrm{D_M/D_H}$', fontsize=24)
ax.set_xlabel(r'redshift z',fontsize=24)
ax.set_xticks([0,1,2,3,4,5])
# ax.yaxis.set_major_locator(MultipleLocator(0.1))
# ax.yaxis.set_major_formatter(NullFormatter())
# ax.set_yticklabels(['','','$0.1$','','$0.3$','','$0.5$','','$0.7$'])
# ax.axes.ticklabel_format(style='sci')
ax.set_ylim(0,3)
ax.set_xlim(0,5)
ax.minorticks_on()
ax.tick_params( labelsize=24)
ax.tick_params('both', length=5, width=1.8,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=8, width=1.8,which='major', direction='in', right='on',top='on')

ax.legend(fontsize=21,frameon=False, loc=2)

plt.show()
```

### Luminosity distance

```python
DL1 = ccl.luminosity_distance(cosmo1,a)
DL2 = ccl.luminosity_distance(cosmo2,a)
DL3 = ccl.luminosity_distance(cosmo3,a)
DL4 = ccl.luminosity_distance(cosmo4,a)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
fig.subplots_adjust(bottom=0.12, left =0.16, right = 0.95, top = 0.95)


ax.plot(z ,DL2/DH,'-', lw=1, color='k', label='E-dS')
ax.plot(z ,DL3/DH,':', lw=1, color='k', label='low density')
ax.plot(z ,DL4/DH,'--', lw=1, color='k', label='high lambda')
ax.plot(z ,DL1/DH,'-.', lw=1, color='k', label='PlanckII')

ax.set_ylabel(r'proper motion distance $\mathrm{D_L/D_H}$', fontsize=24)
ax.set_xlabel(r'redshift z',fontsize=24)
ax.set_xticks([0,1,2,3,4,5])
# ax.yaxis.set_major_locator(MultipleLocator(0.1))
# ax.yaxis.set_major_formatter(NullFormatter())
# ax.set_yticklabels(['','','$0.1$','','$0.3$','','$0.5$','','$0.7$'])
# ax.axes.ticklabel_format(style='sci')
ax.set_ylim(0,16)
ax.set_xlim(0,5)
ax.minorticks_on()
ax.tick_params( labelsize=24)
ax.tick_params('both', length=5, width=1.8,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=8, width=1.8,which='major', direction='in', right='on',top='on')

ax.legend(fontsize=21,frameon=False, loc=2)

plt.show()
```

### Distance Modulus

```python
DM1 = ccl.distance_modulus(cosmo1,1./(z[1:]+1))
DM2 = ccl.distance_modulus(cosmo2,1./(z[1:]+1))
DM3 = ccl.distance_modulus(cosmo3,1./(z[1:]+1))
DM4 = ccl.distance_modulus(cosmo4,1./(z[1:]+1))
```
